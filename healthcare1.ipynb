```python
#no of days patients was hospitalized
df['Days Hospitalized'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
df['Days Hospitalized'] = df['Days Hospitalized'].astype(int)
df['Days Hospitalized']
df
```

```python
# What is the total billing amount for top 10 hospitals?
# you are trying to access a column named 'Hospital' in the top_10 Series,
#  but it doesn't exist as a column. The top_10 Series likely has 'Hospital'
#  as its index, not a column, due to the groupby and sort_values operations.
top_10=df.groupby('Hospital')['Billing_Amount'].sum().sort_values(ascending=False)[:10]
top_10=top_10.reset_index()
top_10
```

```python
sns.barplot(x=top_10['Hospital'],y=top_10['Billing_Amount'])
plt.xticks(rotation=90)
plt.title('Top 10 hospital by billing amount')
plt.show()
```

```python
# Who are the top 20 doctors by billing amount?
top_20=df.groupby('Doctor')['Billing_Amount'].sum().sort_values(ascending=False)[:20]
top_20=top_20.reset_index()
top_20
```

```python
sns.barplot(x=top_20['Doctor'],y=top_20['Billing_Amount'])

plt.xticks(rotation=90)
plt.title('Top 20 doctors by billing amount')
plt.show()
```

```python
# # Which medical condition generated the highest average billing among male and female?
# Grouping data by medical condition and gender then calculating the mean of billing amount
medical_gender_billing = df.groupby(['Medical Condition','Gender'])[['Billing_Amount']].mean().reset_index()
medical_gender_billing
```

```python
sns.barplot(x= medical_gender_billing['Medical Condition'],
            y= medical_gender_billing['Billing_Amount'],
            hue = medical_gender_billing['Gender'],ci = None,
            palette="Accent")
plt.title(f'Barplot of Medical Condition vs Billing Amount')
plt.xticks(rotation=90)
plt.show()
```

```python
# 1. Room Allocation: Count the number of patients in each room
room_allocation = df['Room Number'].value_counts().reset_index()
room_allocation.columns = ['Room Number', 'Patient Count']
room_allocation
```

```python
# 1. Admission Type Analysis
admission_type_counts = df['Admission Type'].value_counts()
admission_type_counts
```

```python
# Grouping by Admission Type and Medical Condition
admission_condition = df.groupby(['Admission Type', 'Medical Condition']).size().unstack().fillna(0)
admission_condition
```

```python
# 2. Seasonal Trends in Admissions
df['Month'] = df['Date of Admission'].dt.month
df['Month']
```

```python
# Group by Month and Medical Condition
monthly_admissions = df.groupby(['Month', 'Medical Condition']).size().unstack().fillna(0)
monthly_admissions
```

```python
#  1. Cost of Care by Condition
cost_by_condition = df.groupby('Medical Condition')['Billing_Amount'].mean().sort_values(ascending=False).reset_index()
cost_by_condition
```

```python
# # 2. Billing vs. Insurance
insurance_billing = df.groupby('Insurance Provider')['Billing_Amount'].mean().sort_values(ascending=False).reset_index()
insurance_billing
```

```python
# # 3. Insurance Provider and Medical Condition Cost Analysis
condition_insurance_cost = df.groupby(['Insurance Provider', 'Medical Condition'])['Billing_Amount'].mean().unstack().fillna(0)
condition_insurance_cost
```

```python
# Condition and Test Results: Analyzing how certain medical conditions relate to test results
# Rows = Unique Medical Conditions.
# Columns = Unique Test Results (Normal, Abnormal, Inconclusive).
# Values = Count of patients for each Medical Condition and Test Result combination.

condition_test_results = df.groupby(['Medical Condition', 'Test Results']).size().unstack().fillna(0)
condition_test_results
```

```python
# Blood Type and Medical Conditions: Investigate whether certain blood types are associated with specific medical conditions
blood_condition = df.groupby(['Blood Type', 'Medical Condition']).size().unstack().fillna(0)
blood_condition
```
